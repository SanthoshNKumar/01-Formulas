```python
'''
Decision Tree Entropy : 

    - Entropy is a measure of the randomness(disorder) in the information being processed
    - The higher the entropy, the harder it is to draw any conclusions from that information
```

```python
'''
Decision Tree Information Gain:
    - Information gain can be defined as the amount of information gained about a random variable or signal from observing 
      another random variable
    - It can be considered as the difference between the entropy of parent node and weighted average entropy of child nodes
    - An attribute with the highest information gain from a set should be selected as the parent (root) node
```

```python
'''
Decision Tree Gini Impurity
    - Gini impurity is a measure of how often a randomly chosen element from the set would be incorrectly labeled if it was 
      randomly labeled according to the distribution of labels in the subset
```

```python
'''
K-Nearest Neighbors (KNN):
    - KNN algorithms use data and classify new data points based on similarity measures (e.g. distance function)
    - Classification is done by a majority vote to its neighbors
    - The data is assigned to the class which has the nearest neighbors.
    - As you increase the number of nearest neighbors, the value of k, accuracy might increase
    
Step-1: Select the number K of the neighbors
Step-2: Calculate the Euclidean distance of K number of neighbors
Step-3: Take the K nearest neighbors as per the calculated Euclidean distance.
Step-4: Among these k neighbors, count the number of the data points in each category.
Step-5: Assign the new data points to that category for which the number of the neighbor is maximum.
Step-6: Our model is ready.
```

```python
'''
XG Boost (Extreme Gradient Boosting)
    - XGBoost is a decision-tree-based ensemble Machine Learning algorithm that uses a gradient boosting framework
    - It is an Extention of Gradient Boosting
    - XG Boost supports multiple language Support
    - It is platform free mean it can run on windows,mac,linux
    - It is Integrable with various system (Like Cloud)
    - It famous because speed of processing and performance
    - It is implemented top of Gradient Boosting and its advancement of Gradient Boosting

What makes XG boost run very fast?
    - Parallelization: If you are running on single processor it will use all cores of the CPU and if you are running 
                       on distributed mode it will ensure it will use maximum available computational power 
                       available of distributed system.
    - Cache Optimization: XG boost keeps all its intermidiate calculation or statistics in cache in such a way that whenever 
                          you try to get the information it will be fast.
    - Out of Memory Computation: Lets say we have data which size is bigger than size of memory. in XG boost there is possible 
                         way of optimizing memory in such a way that it can work on the data which is larget than the RAM.
                         
    - Regularization paramerers in XG boost which avoids overfitting which were not there in normal Gradient Boosting.
      in XG Boost Automatic regularization will happen and makes sure model will not overfitting
    - Auto pruning: in XG boost it will not allow tree to grow after certain level becasue of this model will be 
      effective and robust. this will happen internally by XG Boost automatically
    - XG boost also takes care of Missing values
```

```python
'''
How Auto Pruning happens in XG Boost?


Lambda():(Regularization Parameter): More the value of lambda its an aggressive approach pruning the tree.
Incresing the value of lambda impact of outliers on the prediction will come down significantly

gamma(): More gamma value means we are aggressing about pruning(Splitting trees futher),lesser gamma value means lesser
pruning.

```
